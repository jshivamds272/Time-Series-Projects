```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys,os
import warnings 
warnings.filterwarnings('ignore')
% matplotlib inline
```

```python
np.random.seed(42)
time=np.arange(100)
stationary=np.random.normal(loc=0,scale=1,size=len(time))
```

```python
def run_sequence_plot(x,y,title,xlabel='time',ylabel='series'):
    plt.plot(x,y,'k-')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(alpha=0.3)
```

```python
run_sequence_plot(time,stationary,title='Statioanry TS')
```

```python
#common smoothing technique
#1- Simple smoothing
stationary_time_series_avg=np.mean(stationary)
sts_avg=np.full(shape=len(time),fill_value=stationary_time_series_avg,dtype='float')
```

```python
run_sequence_plot(time,stationary,title='Stationary Data v/s mean')
plt.plot(time,sts_avg,'r',label='mean')
plt.legend()
```

```python
def mse(observations,estimates):
    assert type(observations)==type(np.array([])),
    assert type(estimates)==type(np.array([])),
    assert type(observations)==len(estimates)

    difference=observations-estimates
    sq_diff=difference ** 2
    mae=sum(sq_diff)
    
    return mae
```
