```python
import numpy as np
import sys
import statsmodels as ss
import matplotlib.pyplot as plt
import os
import seaborn as sns

#from colorsetup import colors,palette

% matplolib inline
```

```python
time=np.arange(1,51)
```

```python
trend=time* 2.75
```

```python
plt.plot(time,trend,'b.')
plt.title('Trend vs Time')
plt.xlabel('minute')
plt.ylabel('Electricity demand')
```

```python
seasonal=10*np.sin(time)*10
```

```python
plt.plot(time,seasonal)
plt.title('Seasonal')
plt.xlabel('minutes')
plt.ylabel('electricity demand')
```

```python
np.random.seed(10)
residual=np.random.normal(loc=0,scale=1,size=len(time))
```

```python
plt.plot(time,residual,'r-.')
plt.title('Residual vs time')
plt.xlabel('minutes')
plt.ylabel('electricity demand')
```

aggregate trend,seasonality,residual components

```python
additive=trend+seasonal+residual
```

```python
plt.plot(time,additive,'k.')
plt.title('Additive Time Series')
plt.xlabel('minutes')
plt.ylabel('electricity demand')
```

```python
#multipicative time series
ignored_residual=np.ones_like(residual)
multiplicative=trend*seasonal*ignored_residual
```

```python
plt.plot(time,multiplicative)
plt.title('Multiplicative ')
plt.xlabel('time')
plt.ylabel('electricity')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

ss_decompose=seasonal_decompose(x=additive,model='additive', freq=6)
estimated_trend=ss_decompose.trend
estimated_seasonal=ss_decompose.seasonal
estimated_residual=ss_decompose.resid
```

```python
fig,axes=plt.subplots(4,1,sharex=True,sharey=False)
fig.set_figheight(10)
fig.set_figwidth(15)

axes[0].plot(additive,label='Original')
axes[0].legend(loc='upper left')

axes[1].plot(estimated_trend,label='Trend')
axes[1].legend(loc='upper left')

axes[2].plot(estimated_seasonal, label='Seasonality')
axes[2].legend(loc='upper left')

axes[3].plot(estimated_residual,label='Residuals')
axes[3].legend(loc='upper left')
```

```python
#multiplicative decomposition
ss_decomposition=seasonal_decompose(x=multiplicative,model='multiplicative',freq=6)
estimated_trend=ss_decomposition.trend
estimated_seasonal=ss_decomposition.seasonal
estimated_residual=ss_decomposition.resid
```
